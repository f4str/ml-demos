# Clustering

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
```

## Helper Functions

```python
def cluster(model, X, y):
    reduction = PCA(n_components=3)
    X_reduced = reduction.fit_transform(X)

    fig = plt.figure(figsize=(15, 15))
    ax1 = fig.add_subplot(1, 2, 1, projection='3d')
    model.fit(X_reduced)
    labels = model.labels_
    ax1.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=labels.astype(float), edgecolor='k')
    ax1.set_title('Clustering Algorithm')

    ax2 = fig.add_subplot(1, 2, 2, projection='3d')
    ax2.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y, edgecolor='k')
    ax2.set_title('Ground Truth')

    plt.show()
```

## Load Data

```python
iris = load_iris()
X = iris.data
y = iris.target

print('Shape of X:', X.shape)
print('Shape of y:', y.shape)
```

## K-Means Clustering

```python
kmeans = KMeans(n_clusters=3)
cluster(kmeans, X, y)
```

## DBSCAN Clustering

```python
dbscan = DBSCAN()
cluster(dbscan, X, y)
```
